```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset from Seaborn
df= sns.load_dataset("tips")

# Show the first few rows of the dataset
print(df)
```

```python
ax = sns.countplot(x='sex', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
smokers_df = df[df['smoker'] == 'Yes']
smokers_df

sns.countplot(x='sex', data=smokers_df)
```

```python
smokers_df = df[df['smoker'] == 'Yes']
smokers_df

ax = sns.countplot(x='sex', data=smokers_df)
print(ax)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
plt.plot(df['total_bill'], df['tip'], marker='o', linestyle='-', color='b')
plt.title("Line Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.hist(df['total_bill'], bins=10, color='g', edgecolor='black')
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Seaborn Functions:
# sns.scatterplot(): Creates a scatter plot for visualizing relationships between two variables.
# sns.barplot(): Creates a bar plot to visualize categorical data and their averages.
# sns.boxplot(): Visualizes the distribution of data, including median and outliers.
# sns.heatmap(): Creates a heatmap to visualize correlation or other 2D data.
# Matplotlib Functions:
# plt.plot(): Creates a line plot.
# plt.hist(): Creates a histogram to show data distribution.
# plt.title(), plt.xlabel(), plt.ylabel(): Set the title and axis labels of the plot.
# plt.show(): Displays the plot on the screen.
```
