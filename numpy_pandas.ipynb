```python
import numpy as np
import pandas as pd
monthly_sales = np.random.randint(1000, 5000, size=12)
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
sales_data = pd.DataFrame({
    'Month': months,
    'Sales': monthly_sales
})
sales_data
```

```python
total_sales = sales_data['Sales'].sum()
print(f"Total Sales: {total_sales}")
```

```python
average_sales = sales_data['Sales'].mean()
print(f"Average Sales: {average_sales}")
```

```python
high_sales = sales_data[sales_data['Sales'] > 3000]
print("Months with High Sales:")
print(high_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(sales_data['Month'], sales_data['Sales'], marker='o', linestyle='-', color='b')
plt.title("Monthly Sales", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(True)
plt.show()
```
